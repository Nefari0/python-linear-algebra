```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open('einstein.jpg')
ein = np.mean(ein,2) # average over third dimension (to cancel it)
print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
# testing cell above
print(np.shape(ein))
```

```python
U,s,V = np.linalg.svd(ein) # s is vector, not matrix



fig,ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title('U')

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title('$\\Sigma$')

ax[2].imshow(V)
ax[2].set_title('V')
```

```python
plt.plot(s,'ks-',markerfacecolor='w')
plt.xlim([-1,50])
plt.title('Eigenspectrum of Einstein')
plt.show()
```

### Image Reconstruction

```python
# reshape s matrix for matrix multiplications

print('ein',np.shape(ein))
print('u',np.shape(U))
S = np.zeros(np.shape(ein))
print('S',np.shape(S))
for i in range(0,len(s)):
    S[i,i] = s[i]
#     print(S[i,i])

r_ein = U@S@V

print(np.shape(r_ein))
```

```python
# display images

fig,ax = plt.subplots(1,2)

ax[0].imshow(ein)
ax[0].set_title('original')


ax[1].imshow(r_ein)
ax[1].set_title('SVD version')

plt.show()
```
