```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# experimental
M = np.array([[.5,1], 
              [1,.5]])

v = np.array([1,1]) # defined
v = np.random.randn(2) # random


Mv = M@v

plt.plot([0,v[0]],[0,v[1]],linewidth=2,label='v') # plot v
plt.plot([0,Mv[0]],[0,Mv[1]],'r--',linewidth=2,label='Mv') # plot Mv

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
# visualization prototype

M = np.array([[.5,1],[1,.5]])
a = np.linspace(-2,2,30)

for i in a:
    v = np.array([1,i])
    Mv = M@v
    plt.plot([0,v[0]],[0,v[1]])
    plt.plot([0,Mv[0]],[0,Mv[1]]) # plot Mv
```
