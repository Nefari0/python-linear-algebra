```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
# experimental/concepts
X = np.array([ [2,1,-1],[3,4,2],[1,-5,-2] ])
y = np.array([1,13,0])

beta1 = np.linalg.inv(X.T@X) @ (X.T@y)
beta2 = np.linalg.solve(X,y)
print(beta1)
print(beta2)
```

### Exercise. Instructed

```python
X1 = np.array([ [3,-1],[-1,1] ])
y1 = np.array( [6,2] )


X2 = np.array([ [3,-1],[1.5,-.5] ])
y2 = np.array( [6,3] )

X3 = np.array([ [3,-1],[1.5,-.5] ])
y3 = np.array( [6,2] )
print()
```

```python
#  system 1)

xlim = np.array([0,10])

#  convert to y = mx + b format
yy1 = -X1[0,0]/X1[0,1]*xlim + y1[0]/X1[0,1]
yy2 = -X1[1,0]/X1[1,1]*xlim + y1[1]/X1[1,1]

plt.plot(xlim,yy1,'b',label="eq 1")
plt.plot(xlim,yy2,'r',label="eq 2")

# -- solve the system -- #
b = np.linalg.solve(X1,y1)
print(b)
plt.plot(b[0],b[1],'ko',label="solution")

plt.legend()
plt.grid()
```

```python
#  system 2)

# this system has infinite solution 

xlim = np.array([0,10])

#  convert to y = mx + b format
yy1 = -X2[0,0]/X2[0,1]*xlim + y2[0]/X2[0,1]
yy2 = -X2[1,0]/X2[1,1]*xlim + y2[1]/X2[1,1]

plt.plot(xlim,yy1,'b',label="eq 1")
plt.plot(xlim,yy2,'r--',label="eq 2")

# -- solve the system -- #
# this system cannot be solved
# b = np.linalg.solve(X2,y2)
# print(b)
# plt.plot(b[0],b[1],'ko',label="solution")

plt.legend()
plt.grid()
```
